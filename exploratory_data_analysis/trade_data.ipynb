```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
trade_stats = pd.read_csv("F:/my files/python files/neural networks/datasets/commodity_trade_statistics_data.csv")
#trade_stats = trade_stats.fillna('nil')
trade_stats.head()
```

```python
trade_stats.columns
```

```python
ind = trade_stats[trade_stats['country_or_area']=='India']
ind
```

```python
ind_export = ind[ind['flow']=='Export']
ind_export
```

```python
ind_exp = ind_export.groupby(['year'], as_index=False)['weight_kg','quantity'].agg('sum')
ind_exp
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(20,10))
sns.barplot(x='year', y='weight_kg', data=ind_exp, palette='gist_rainbow')
```
