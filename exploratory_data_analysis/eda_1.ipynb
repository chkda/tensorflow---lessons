```python
#scatter plot using seaborn
sns.set_style('whitegrid')
sns.FacetGrid(iris, hue='species', size=5)\
    .map(plt.scatter, 'sepal_length', 'sepal_width')\
    .add_legend()
    
plt.show()
```

```python
sns.set_style('whitegrid')
sns.pairplot(iris, hue='species', size=3)
plt.show()
```

## Cumulative Distribution Function

```python
iris_setosa = iris[iris['species']=='setosa']
iris_setosa.head()
```

```python
iris_versi = iris[iris['species']=='versicolor']
iris_versi.head()
```

```python
iris_virgin = iris[iris['species']=='virginica']
iris_virgin.head()
```

```python
counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins=10, density=True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

counts1, bin_edges1 = np.histogram(iris_versi['petal_length'], bins=10, density=True)
pdf1 = counts1/(sum(counts1))
cdf1 = np.cumsum(pdf1)
plt.plot(bin_edges1[1:], pdf1)
plt.plot(bin_edges1[1:], cdf1)

counts2, bin_edges2 = np.histogram(iris_virgin['petal_length'], bins=10, density=True)
pdf2 = counts/(sum(counts))
cdf2 = np.cumsum(pdf2)
plt.plot(bin_edges2[1:], pdf2)
plt.plot(bin_edges2[1:], cdf2)
```
